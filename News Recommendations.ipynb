```python
urllist=[]
for i in range(1,6,1):
    url="https://www.opindia.com/latest-news/page/" + str(i)
    urllist.append(url)
```

```python
urllist
```

```python
headlines=[]
import requests
from bs4 import BeautifulSoup
for j in urllist:
    data=requests.get(j)
    soup=BeautifulSoup(data.content,"html.parser")
    for i in soup.find_all("h3",class_="entry-title td-module-title"):
        headlines.append(i.find("a")["title"])
```

```python
headlines
```

# Preprocessing of Text data

```python
ppd=[]    #preprocessing data
import re
for i in headlines:
    headlines=i.lower()
    headlines=re.sub("[^a-zA-Z0-9 ]","",headlines)
    ppd.append(headlines)
```

```python
ppd
```

```python
len(ppd)
```

# Stemming and Lemmatization

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()

from nltk.stem import WordNetLemmatizer 
lem = WordNetLemmatizer()
E = []
ppdf = []
for j in range(0,len(ppd)):
    wnl = []
    for i in ppd[j].split(" "):
        wnl.append(lem.lemmatize(ls.stem(i)))
        E.append(lem.lemmatize(ls.stem(i)))
        
    ppdf.append(wnl)
```

```python
ppdf
```

# Cluster Prepration

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(E).toarray()
```
